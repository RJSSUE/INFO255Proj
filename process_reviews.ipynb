```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/reviews.csv')
df.head()
```

```python
df.shape
```

```python
df_no_dup = df.drop_duplicates(subset='review_id', keep="first")
df_no_dup.shape
```

```python
df_no_dup['user_first_name'] = df_no_dup.user_name.map(lambda x: x.split(' ')[0])
```

```python
df_no_dup.head()
```

```python
df_no_dup.to_csv(r'./data/reviews_with_first_name.csv', index=False)
```
